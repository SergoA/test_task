```python
import pandas as pd

# Открываем файл
df = pd.read_csv('Big_Black_Money_Dataset.csv')

# Преобразование столбца 'Date of Transaction' в тип datetime
df['Date of Transaction'] = pd.to_datetime(df['Date of Transaction'])

# Проверяем, что 'Date of Transaction' является индексом для использования resample
df.set_index('Date of Transaction', inplace=True)

# Агрегация по каждому часу
hourly_aggregation = df.resample('H').sum()

# Преобразуем индекс в столбец для объединения
hourly_aggregation.reset_index(inplace=True)

# Присоединяем метрики ближайшего часа
df.reset_index(inplace=True)
df = df.merge(hourly_aggregation, how='left', on='Date of Transaction', suffixes=('', '_hourly'))

# выводим результат
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Открываем файл
df = pd.read_csv('Big_Black_Money_Dataset.csv')

# Преобразование столбца 'Date of Transaction' в тип datetime
df['Date of Transaction'] = pd.to_datetime(df['Date of Transaction'])

# Отрисовываем график для столбца Amount (USD)
numeric_column = 'Amount (USD)'

# Устанавливаем индекс по дате
df.set_index('Date of Transaction', inplace=True)

# Группировка по месяцам и вычисление среднего значения
monthly_mean = df.resample('M')[numeric_column].mean()

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(monthly_mean.index, monthly_mean, marker='o', color='blue', alpha=0.7)
plt.title(f'Среднее значение {numeric_column} по месяцам')
plt.xlabel('Месяц')
plt.ylabel(f'Среднее значение {numeric_column}')
plt.grid()
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Открываем файл
df = pd.read_csv('Big_Black_Money_Dataset.csv')

# Замените 'Transaction Type' на имя вашей колонки
string_column = 'Transaction Type'

# Подсчет частоты каждого символа
char_frequency = {}

for entry in df[string_column].dropna():
    for char in entry:
        if char in char_frequency:
            char_frequency[char] += 1
        else:
            char_frequency[char] = 1

# Преобразование в DataFrame
char_freq_df = pd.DataFrame(char_frequency.items(), columns=['Character', 'Frequency'])

# Создание сводной таблицы для heatmap
heatmap_data = char_freq_df.set_index('Character')

# Построение тепловой карты
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt='g', cmap='Blues', cbar=True)
plt.title('Тепловая карта частотности символов в колонке "Transaction Type"')
plt.xlabel('Частота')
plt.ylabel('Символы')
plt.show()
```
