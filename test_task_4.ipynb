```python
import pandas as pd
import swifter

   # Открываем файл на компьютере
df = pd.read_csv('Big_Black_Money_Dataset.csv')

   # просмватриваем информацию о датасете и типе данных по столбцам
type (df)
df.info()
```

```python
#преобразуем столбец даты в тип datetime
df = pd.to_datetime(df['Date of Transaction'])

df.info()
```

```python
#чистим датасет и удаляем строки где есть хотябы одно пустое значиние
df_cleaned = df.dropna()
df_cleaned.info()
```

```python
#удаляем дубли
df_unique = df_cleaned.drop_duplicates()
df_unique.info()

#выбираем уникальные значения по столбцу и сортируем их по возрастанию
unique_values = df_unique['Financial Institution'].unique()


#Сортируем данные чтобы увидеть, что первым стоит значение из букв
sorted_unique_values = sorted(unique_values)
print (sorted_unique_values)

df_unique.info()
```

```python
#строки в которых нет цифр превращаем в пустые

# Функция для проверки наличия цифр
def replace_if_no_digits(s):
    # Проверяем, есть ли цифры в строке
    if any(char.isdigit() for char in s):
        return s
    else:
        return ''

# Применяем функцию к столбцу 'Financial Institution', используем .loc для избежания SettingWithCopyWarning
df_unique.loc[:, 'Financial Institution'] = df_unique['Financial Institution'].apply(replace_if_no_digits)


#образец кода с применением распараллеливания выполнения шагов:
#df_unique['Financial Institution'] = df_unique['Financial Institution'].swifter.apply(replace_if_no_digits)


# Просмотр обновленного DataFrame
df_unique
```

```python
#Удаление записей по временному промежутку
df_unique.info()

# Удаление записей в промежутке от 1 до 3 часов ночи
df_filtered = df_unique[(df_unique['Date of Transaction'].dt.hour < 1) | (df_unique['Date of Transaction'].dt.hour >= 3)]

df_filtered
```
